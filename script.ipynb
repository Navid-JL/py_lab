```python
img = Image.open('man.jpg')
```

```python
import numpy as np
```

```python
pic_array = np.array(img)
```

```python
pic_array
```

```python
mask = np.full(pic_array.shape, 255)
```

```python
mask
```

```python
modified_array = pic_array - mask

modified_array = modified_array * -1

modified_array = modified_array.astype(np.uint8)
```
